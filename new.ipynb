```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df= pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='red')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d= pd.read_csv('areas.csv')
d.head(3)
```

```python
p =reg.predict(d)
```

```python
d['prices']=p
d
```

```python
d.to_csv('predicted.csv',index=False)
```

```python
pr=pd.read_csv('predicted.csv')
pr
```

```python
%matplotlib inline
plt.scatter(pr.area, pr.prices,color='green')
plt.plot(pr.area, pr.prices,color='red')
```

```python
import pickle
with open('model_pickle','wb')as f:
    pickle.dump(reg,f)
```
